```python
import pandas as pd
import math
pd.set_option('display.max_rows', None)
```

```python
Shark_path = "Resources/attacks.csv"
```

```python
Shark_df = pd.read_csv(Shark_path, encoding="ISO-8859-1")
Shark_df.head()
```

```python
# Drop the columns of the data that will not be used
Shark_df = Shark_df.drop(columns={"Case Number","Date","Type","Area","Location","Name","Age","Species ","Injury","Time","Investigator or Source","pdf","href formula","href","Case Number.1","Case Number.2","original order","Unnamed: 22","Unnamed: 23"})
Shark_df.head()
```

```python
Shark_df.rename(columns = {"Fatal (Y/N)":"Fatal"}, inplace = True)
Shark_df.head()
```

```python
Shark_df.dtypes
```

```python
cleanshark_df = Shark_df.fillna (0)
cleanshark_df.tail()
```

```python
cleanshark_df ['Year'] = cleanshark_df['Year'].fillna(0).astype(int)
cleanshark_df.head()
```

```python
cleanshark_year = cleanshark_df[cleanshark_df["Year"] >= 1970]
cleanshark_year.head()
```
